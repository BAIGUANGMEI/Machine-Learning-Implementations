```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/basic/Employers_data.csv')
df.head()
```

```python
# Line Graph
df.groupby('Department')['Salary'].mean().plot(kind='line', title='Average Salary by Department')
plt.show()
```

```python
# Bar Chart
df.groupby('Department')['Name'].count().plot(kind='bar', title='Number of Employees by Department')
plt.show()
```

```python
# Histogram
df['Salary'].plot(kind='hist', bins=10, title='Salary Distribution')
plt.show()
```

```python
# scatter Plot
df.plot(kind='scatter', x='Age', y='Salary', title='Age vs Salary')
plt.show()
```

```python
# box Plot
df.plot(kind='box', column='Salary', by='Department')
plt.title('Salary Distribution by Department')
plt.suptitle('')  # Remove the default suptitle
plt.show()
```

```python
# pie Chart
df.groupby('Department')['Name'].count().plot(kind='pie', title='Employee Distribution by Department', autopct='%1.1f%%')
plt.ylabel('')  # Hide y-label
plt.show()
```
