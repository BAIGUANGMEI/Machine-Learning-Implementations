```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'], name='MySeries')
print(s)
```

```python
s1 = pd.Series({'apple': 3, 'banana': 5, 'orange': 2}, name='FruitCounts')
print(s1)
```

```python
print(s1.index)
print(s1.values)
print(s1.dtype)
print(s1.shape)
print(s1.size)
print(s1.ndim)


print(s1.loc['banana'])
print(s1.iloc[1])
print(s1.at['banana'])
print(s1.iat[1])
```

```python
print(s.head(3))
print(s.tail(2))
print(s.isin([20, 40]))
print(s.isna())
print(s.describe())
```

```python
series = pd.Series([100, 200, 300, 400, np.nan, None,600], index=['a', 'b', 'c', 'd', 'e', 'f','g'])
print(series)
```

```python
print(series.describe())
print(series.isna())
print(series.isin([200, np.nan]))
```

```python
print(series.value_counts())
print(series.count())
```
